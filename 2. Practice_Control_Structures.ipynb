- **Cüt və ya Tək Rəqəm Yoxlanışı (If/Else)
Tapşırıq**
-check_even_odd(num) adlı bir funksiya yaradın:
- Parametr kimi bir tam ədəd qəbul etsin.
- If/Else vasitəsi ilə ədədin cüt və ya tək olduğunu yoxlasın.
- Cüt olarsa "Cüt", tək olarsa "Tək" qaytarsın (və ya çap etsin)



```python

def check_even_odd(num):

  if num%2==0:
    value= 'Cüt'
  else:
     value ='Tek'
  return value
# Funksiyanı sınaqdan keçirmək
test_number = 37
result = check_even_odd(test_number)
print(f"{test_number} ədədi {result}-dir")
```

2.**while Dövrü ilə Toplama**
- sum_upto_n(n) adlı funksiya yazın:
- 1-dən n-ə qədər (n daxil olmaqla) bütün ədədləri toplayın (bunun üçün while dövründən istifadə edin).
- Toplamı qaytarsın.
- n mənfi olarsa, 0 qaytarın və ya xəbərdarlıq mesajı çap edin (öz seçiminizdir).

```python
def sum_upto_n(n):

    pass
    if n < 0:
        print("Xəta: Mənfi ədəd daxil etdiniz!")

    total = 0
    i = 1
    while i <= n:
        total += i
        i += 1

    return total

# Test
test_values = [5, 10, -3, 0]
for val in test_values:
    print(f"sum_upto_n({val}) = {sum_upto_n(val)}")
```

3. **FizzBuzz**
- fizz_buzz(n) adlı bir funksiya yaradın və 1-dən n-ə qədər dövr edin:
- Əgər ədəd 3-ə bölünürsə, "Fizz",
- Əgər ədəd 5-ə bölünürsə, "Buzz",
- Əgər ədəd həm 3-ə, həm də 5-ə bölünürsə, "FizzBuzz",
- Əks halda, ədədin özünü çap edin.

```python
def fizz_buzz(n):

    for i in range(1, n + 1):  # 1-dən n-ə qədər dövr
        if i%3==0:
          print('Fizz')
        elif i % 5 == 0:
            print("Buzz")
        elif i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        else:
            print(i)

# Test
fizz_buzz(15)
```

**Saitlərin Sayını Tapmaq**
- count_vowels(text) adlı funksiya yazın:
- Giriş kimi sətir (string) qəbul etsin.
- Saitlərin (a, e, i, o, u və eyni zamanda böyük hərflər) sayını - hesablamaq üçün dövrdən istifadə etsin.


```python

def count_vowels(text):
  text=text.lower()
  vowels = "aeiou"        # həm böyük, həm kiçik saitlər
  count = 0               # sayğac sıfırdan başlayır
  for i in text:          # hər bir hərfi yoxlayırıq
     if i in vowels:      # əgər hərf saitdirsə
        count += 1
  return count


# Test
test_strings = ["Hello", "Education", "Rhythm", ""]
for t in test_strings:
    print(f"'{t}' sözündə saitlərin sayı: {count_vowels(t)}")
```
